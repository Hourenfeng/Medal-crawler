```python
###运行整个程序的代码包
from bs4 import BeautifulSoup
from datetime import datetime
import pandas
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import jieba
import requests
import re
from selenium import webdriver
from selenium.webdriver.common.action_chains import ActionChains
import csv
import time
import codecs
import numpy as np
```

```python
###爬取网页信息并存为EXCEL文件
driver=webdriver.Chrome()
url='http://match.2016.sina.com.cn/medals?cre=bd.ala.ayjpb'
driver.get(url)

xpath='//*[@id="html"]/body/div[1]/div[2]/div[2]'
tabel=driver.find_element_by_xpath(xpath).get_attribute('innerHTML')


soup=BeautifulSoup(tabel,"html.parser")
tabel1=soup.find_all('tr')
i=0
lst=[]
biaoti=['国家','金牌','银牌','铜牌','总计']
f=codecs.open('jiangpai.csv','w','gbk')
writer=csv.writer(f)
writer.writerow(biaoti)
f.close()
for row in tabel1:
    cols=[col.text for col in row.find_all('td')]
    newcols=[x.strip() for x in cols if x.strip()!='']
    if newcols!=lst:
        if newcols[0].isdigit():
            del newcols[0]
            t=codecs.open('jiangpai.csv','a','gbk')
            writer=csv.writer(t)
            writer.writerow(newcols)
            print(newcols)
            time.sleep(1)
            t.close()
```

```python
###数据可视化
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus'] = False

f=open('jiangpai.csv','r+',encoding='gbk')###将此文件放在默认文件夹下面
csvreader=csv.reader(f)
guojia=[row[0] for row in csvreader]
del guojia[0]
f.close()
t=open('jiangpai.csv','r+',encoding='gbk')
reader=csv.reader(t)
jinpai=[row[1] for row in reader]
del jinpai[0]
t.close()
t=open('jiangpai.csv','r+',encoding='gbk')
reader=csv.reader(t)
yinpai=[row[2] for row in reader]
del yinpai[0]
t.close()
t=open('jiangpai.csv','r+',encoding='gbk')
reader=csv.reader(t)
tongpai=[row[3] for row in reader]
del tongpai[0]
t.close()

jinpai=[int(x) for x in jinpai]
jin=np.array(jinpai)
yinpai=[int(x) for x in yinpai]
yin=np.array(yinpai)
tongpai=[int(x) for x in tongpai]
tong=np.array(tongpai)
country=guojia[0:7]
jina=np.array(jinpai[0:7])
yina=np.array(yinpai[0:7])
tonga=np.array(tongpai[0:7])

###绘制水平堆叠条形图
fig = plt.figure(figsize=(9,15))
plt.barh(guojia, jin, color='green', label='金牌')
plt.barh(guojia, yin, left=jin, color='red', label='银牌')
plt.barh(guojia, tong, left=jin+yin, color='blue', label='铜牌')
plt.title("Top Seven Countries in Rio")                                                      
plt.xlabel("奖牌数")                                                             
plt.legend(loc=[0, 0])                                         
plt.show()

###绘制饼图
plt.figure(figsize=(6,6))
label=country
explode=[0.01,0.01,0.1,0.01,0.01,0.01,0.01]
values=jina+yina+tonga
plt.pie(values,explode=explode,labels=label,autopct='%1.1f%%')
plt.title('里约奥运会前七名国家')
plt.show()
```
